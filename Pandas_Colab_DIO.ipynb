```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
store_names = [f'Loja_{i}' for i in range(1, 17)]
```

```python
products = [{'ID do Produto': i,
             'Marca': f'Marca_{np.random.randint(1, 11)}',
             'Cor': np.random.choice(['Vermelho', 'Azul', 'Verde', 'Amarelo', 'Preto', 'Branco']),
             'Tamanho': np.random.choice(['P', 'M', 'G', 'GG']),
             'Preço': np.random.uniform(10, 100)} for i in range(1, 51)]
```

```python
df_produtos = pd.DataFrame(products)
```

```python
dados_vendas = {'Loja': np.random.choice(store_names, 1000),
                'ID do Produto': np.random.choice(df_produtos['ID do Produto'], 1000),
                'Mês': np.random.choice(['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho',
                                         'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'], 1000),
                'Vendas': np.random.randint(1, 20, 1000)}
```

```python
df_vendas = pd.DataFrame(dados_vendas)
```

```python
vendas_mensais = df_vendas.groupby('Mês')['Vendas'].sum().sort_values()
```

```python
plt.figure(figsize=(12, 6))
vendas_mensais.plot(kind='bar')
plt.title('Vendas Mensais')
plt.xlabel('Mês')
plt.ylabel('Total de Vendas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
